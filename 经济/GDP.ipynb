```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./GDP growth.csv')
df.keys()[47:-1]
```

```python
label = df.keys()[47:-1]
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
output = imputer.fit_transform(df[label])
```

```python
df[label] = output
```

```python
df = df.drop(df.keys()[2:47], axis=1)
df = df.drop(df.keys()[-1], axis=1)
df = df.drop(df.keys()[0], axis=1)
```

```python
df.to_csv('New_New_GDP growth.csv')
```

```python
label = df.keys()[1:]
```

```python
# 标准化
def standard(datas):
    k = np.power(np.sum(np.power(datas, 2)), 0.5)
    datas = datas / k
    return datas
```

```python
list = []
for value in df[label].values:
    list.append(standard(value))
```

```python
df[label] = list
```
