```python
def interToMax(datas, a, b):
    M = max(a - datas.min(), datas.max() - b)
    ans = []
    for data in datas:
        if data < a:
            ans.append(1 - (a - data) / M)
        elif data > b:
            ans.append(1 - (data - b) / M)
        else:
            ans.append(1)
    return ans
label = df.keys()[1:]
list = []
for value in df[label].values:
    list.append(interToMax(value, 30, 40))
df[label] = list
```

```python
# 标准化
def standard(datas):
    k = np.power(np.sum(np.power(datas, 2)), 0.5)
    datas = datas / k
    return datas
```

```python
label = df.keys()[1:]
list = []
for value in df[label].values:
    list.append(standard(value))
```
