```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./gini index.csv')
df1 = pd.read_csv('./New_unemployment.csv')
lists = df1['Unnamed: 0'].tolist()
filled_df = pd.DataFrame()
for list in lists:
    filled_df = filled_df.append(df[list:list + 1])
df = filled_df
```

```python
label = df.keys()[34:-1]
len(label)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer()
outputs = imputer.fit_transform(df[label].values)
```

```python
df[label] = outputs
```

```python
df = df.drop(df.keys()[1:34], axis=1)
df = df.drop(df.keys()[-1], axis=1)
```

```python
df
```

```python
def interToMax(datas, a, b):
    M = max(a - datas.min(), datas.max() - b)
    ans = []
    for data in datas:
        if data < a:
            ans.append(1 - (a - data) / M)
        elif data > b:
            ans.append(1 - (data - b) / M)
        else:
            ans.append(1)
    return ans
label = df.keys()[1:]
list = []
for value in df[label].values:
    list.append(interToMax(value, 30, 40))
df[label] = list
```

```python
df
```

```python
# 标准化
def standard(datas):
    k = np.power(np.sum(np.power(datas, 2)), 0.5)
    datas = datas / k
    return datas
```

```python
label = df.keys()[1:]
list = []
for value in df[label].values:
    list.append(standard(value))
```

```python
df[label] = list
df
```

```python
df.to_csv('New_gini index.csv')
```
