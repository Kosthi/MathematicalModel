```python
import numpy as np
import pandas as pd
df = pd.read_excel("20条河流的水质情况数据.xlsx")
df.head()
```

```python
def minToMax(datas):
    return datas.max() - datas
df['细菌总数(个/mL)'] = minToMax(df['细菌总数(个/mL)'])
df.head()
```

```python
def midToMax(datas, x_best):
    M = abs(datas - x_best).max()
    return 1 - abs(datas - x_best) / M
df['PH值'] = midToMax(df['PH值'], 7)
df.head()
```

```python
def interToMax(datas, a, b):
    M = max(a - datas.min(), datas.max() - b)
    ans = []
    for data in datas:
        if data < a:
            ans.append(1 - (a - data) / M)
        elif data > b:
            ans.append(1 - (data - b) / M)
        else:
            ans.append(1)
    return ans
df['植物性营养物量（ppm)'] = interToMax(df['植物性营养物量（ppm)'], 10, 20)
df.head()
```

```python
def standard(datas):
    k = np.power(np.sum(np.power(datas, 2), axis = 0), 0.5)
    for i in range(len(k)):
        datas[:, i] = datas[:, i] / k[i]
    return datas
label = df.keys()[1:]
data = standard(df[label].values)
```

```python
def score(datas):
    z_min = np.min(datas, axis = 0)
    z_max = np.max(datas, axis = 0)
    min_Distance = np.power(np.sum(np.power(z_min - datas, 2), axis = 1), 0.5)
    max_Distance = np.power(np.sum(np.power(z_max - datas, 2), axis = 1), 0.5)
    scores = min_Distance / (max_Distance + min_Distance)
    scores = scores / np.sum(scores, axis = 0)
    return scores

df['得分'] = score(data)
df
```
