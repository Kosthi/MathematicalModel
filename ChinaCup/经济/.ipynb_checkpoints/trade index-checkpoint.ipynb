```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./fiscal balance index.csv')
df.keys()[13:]
df
```

```python
label = df.keys()[13:]
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
output = imputer.fit_transform(df[label])
```

```python
df[label] = output
```

```python
df = df.drop(df.keys()[2:13], axis=1)
# df = df.drop(df.keys()[-1], axis=1)
df = df.drop(df.keys()[0], axis=1)
```

```python
df
```

```python
# 标准化
def standard(datas):
    k = np.power(np.sum(np.power(datas, 2)), 0.5)
    datas = datas / k
    return datas
list = []
for value in df[label].values:
    list.append(standard(value))
df[label] = list
```

```python
df
```

```python
df.to_csv('New_fiscal balance index.csv')
```
