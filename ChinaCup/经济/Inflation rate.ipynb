```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./Inflation rate.csv')
df.keys()[47:-1]
```

```python
label = df.keys()[47:-1]
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
output = imputer.fit_transform(df[label])
```

```python
df[label] = output
```

```python
df = df.drop(df.keys()[2:47], axis=1)
df = df.drop(df.keys()[-1], axis=1)
df = df.drop(df.keys()[0], axis=1)
```

```python
df.to_csv('New_New_Inflation rate.csv')
```

```python
label = df.keys()[1:]
list = []
def interToMax(datas, a, b):
    M = max(a - datas.min(), datas.max() - b)
    ans = []
    for data in datas:
        if data < a:
            ans.append(1 - (a - data) / M)
        elif data > b:
            ans.append(1 - (data - b) / M)
        else:
            ans.append(1)
    return ans
for value in df[label].values:
    list.append(interToMax(value, 3, 5))
df[label] = list
```

```python
# 标准化
def standard(datas):
    k = np.power(np.sum(np.power(datas, 2)), 0.5)
    datas = datas / k
    return datas
list = []
for value in df[label].values:
    list.append(standard(value))
df[label] = list
```
