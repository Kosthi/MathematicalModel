```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
```

```python
# 处理投入指标
data_1 = pd.read_csv('./指标/投入：自然资源/gdp.csv', skiprows=[0, 1, 2, 3])
data_2 = pd.read_csv('./指标/投入：自然资源/农业用地.csv', skiprows=[0, 1, 2, 3])
data_3 = pd.read_csv('./指标/投入：自然资源/可再生内陆淡水资源总量.csv', skiprows=[0, 1, 2, 3])
data_4 = pd.read_csv('./指标/投入：自然资源/森林面积.csv', skiprows=[0, 1, 2, 3])
data_5 = pd.read_csv('./指标/投入：自然资源/耕地面积.csv', skiprows=[0, 1, 2, 3])
data_6 = pd.read_csv('./指标/投入：自然资源/陆地及海洋保护区面积.csv', skiprows=[0, 1, 2, 3])
selected = pd.read_csv('./指标/投入：自然资源/gdp copy.csv')
countries = selected['Country Name'].values

def select(data):
    data_cleaned = pd.DataFrame()
    for country in countries:
        for i in range(len(data_1)):
            if country == data.iloc[i, 0]:
                data_cleaned[i] = data.loc[i]
    return data_cleaned
data1_cleaned = select(data_1).T.drop(['Unnamed: 66', '2021'], axis=1)
data2_cleaned = select(data_2).T.drop(['Unnamed: 66', '2021'], axis=1)
data3_cleaned = select(data_3).T.drop(['2021'], axis=1)
data4_cleaned = select(data_4).T.drop(['Unnamed: 66', '2021'], axis=1)
data5_cleaned = select(data_5).T.drop(['Unnamed: 66', '2021'], axis=1)
data6_cleaned = select(data_6).T.drop(['Unnamed: 66', '2021'], axis=1)

labels = data1_cleaned.keys()[34:]
# 填补缺失值
def fill(data_cleaned):
    imoputer = KNNImputer(n_neighbors=2)
    return imoputer.fit_transform(data_cleaned[labels])
data1_cleaned[labels] = fill(data1_cleaned)
data2_cleaned[labels] = fill(data2_cleaned)
data3_cleaned[labels] = fill(data3_cleaned)
data4_cleaned[labels] = fill(data4_cleaned)
data5_cleaned[labels] = fill(data5_cleaned)
data6_cleaned[labels[-5]] = fill(data6_cleaned)
# 去除多余列
data1_cleaned.drop(data1_cleaned.keys()[1:54], axis=1, inplace=True)
data2_cleaned.drop(data2_cleaned.keys()[1:54], axis=1, inplace=True)
data3_cleaned.drop(data3_cleaned.keys()[1:54], axis=1, inplace=True)
data4_cleaned.drop(data4_cleaned.keys()[1:54], axis=1, inplace=True)
data5_cleaned.drop(data5_cleaned.keys()[1:54], axis=1, inplace=True)
data6_cleaned.drop(data6_cleaned.keys()[1:54], axis=1, inplace=True)
```

```python
# 对data6均值填充
label = labels[-11:-5]
for i in range(6):
    data6_cleaned[label[-(i + 1)]] = data6_cleaned[labels[-5 - i:]].mean(axis=1)
```

```python
# 保存
data1_cleaned.to_csv('./指标/投入：自然资源/cleaned_gdp.csv')
data2_cleaned.to_csv('./指标/投入：自然资源/cleaned_农业用地.csv')
data3_cleaned.to_csv('./指标/投入：自然资源/cleaned_可再生内陆淡水资源总量.csv')
data4_cleaned.to_csv('./指标/投入：自然资源/cleaned_森林面积.csv')
data5_cleaned.to_csv('./指标/投入：自然资源/cleaned_耕地面积.csv')
data6_cleaned.to_csv('./指标/投入：自然资源/cleaned_陆地及海洋保护区面积.csv')
```
