```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./Intentional homicides (per 100,000 people).csv')
df1 = pd.read_excel('../指标总.xlsx')
lists = df1['Country Name'].tolist()
filled_df = pd.DataFrame()
k = []
for list in lists:
    flag = True
    for i in range(0, 266):
        if df.loc[i]['Country Name'] == list:
            flag = False
            filled_df = filled_df.append(df.loc[i])
            break
    if flag:
        print(list)
        tmp = df1[:]
        for i in range(0, 121):
            if tmp.loc[i]['Country Name'] == list:
                k.append(i)
df = filled_df
```

```python
df = df.drop(df.keys()[1:46], axis=1)
df = df.drop(df.keys()[-1], axis=1)
```

```python
label = df.keys()[1:]
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
outputs = imputer.fit_transform(df[label].values)
```

```python
df[label] = outputs
```

```python
df.to_csv('New_New_Intentional homicides.csv')
```

```python
def minToMax(datas):
    return datas.max() - datas
label = df.keys()[1:]
list = []
for value in df[label].values:
    list.append(minToMax(value))
df[label] = list
```

```python
# 标准化
def standard(datas):
    k = np.power(np.sum(np.power(datas, 2)), 0.5)
    datas = datas / k
    return datas
```

```python
label = df.keys()[1:]
list = []
for value in df[label].values:
    list.append(standard(value))
```

```python
df[label] = list
df
```
