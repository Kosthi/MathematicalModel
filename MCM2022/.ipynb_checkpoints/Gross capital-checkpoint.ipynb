```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
def missing_values(dataframe):
    missing_ratio = (dataframe.isnull().sum(axis=1) / len(dataframe))*100
    missing_count = dataframe.isnull().sum(axis=1)
    info = pd.DataFrame({'Country': dataframe[dataframe.keys()[0]], 'Missing Ratio': missing_ratio, 'Missing Count': missing_count})
    return info
```

```python
df = pd.read_csv('MCM2022F题数据/经济/Adjusted net national income per capita (current US$).csv')
df1 = pd.read_csv('./Poverty headcount ratio at $2.15 a day.csv')
lists = df1['Unnamed: 0'].tolist()
filled_df = pd.DataFrame()
for list in lists:
    filled_df = filled_df.append(df[list:list + 1])
filled_df
```

```python
#df = df.dropna(how='all', subset=df.keys()[4:66])
filled_df = filled_df.dropna(thresh=3, subset=df.keys()[4:66])
filled_df.count()

m=filled_df[filled_df.keys()[4:66]].mean(axis=1)
filled_df=filled_df.fillna({x:m for x in filled_df.keys()[4:66]})
filled_df
# filled_df.to_csv('Poverty headcount ratio.csv')
```
